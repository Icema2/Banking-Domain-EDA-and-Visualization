```python
#converting numerical to categorical data
bins= [0,100000,300000,float('inf')]
labels=['Low','Mid','High']

df['Income Band']= pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique columns in categorical columns 
# simple univariate analysis 
categorical_cols= df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality',
                      'Occupation','Fee Structure','Loyalty Classification','Properties Owned',
                      'Risk Weighting','Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

# some values are equally distributed while some are not 

# Univariate Analysis

```python
sns.set_style("whitegrid")

for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality',
                                  'Occupation','Fee Structure','Loyalty Classification','Properties Owned',
                                  'Risk Weighting','Income Band']].columns):
    plt.figure(i, figsize=(10, 6))
    sns.countplot(data=df, x=predictor, hue=predictor, palette="muted", legend=False)  # No warning
    plt.title(f'Count Plot of {predictor}', fontsize=14)
    plt.xlabel(predictor, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
```

# Bivariate Analysis

# Analysing with respect to Gender

```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality',
                                  'Occupation','Fee Structure','Loyalty Classification','Properties Owned',
                                  'Risk Weighting','Income Band']].columns):
    plt.figure(i, figsize=(10, 6))  # Set figure size
    sns.countplot(data=df, x=predictor, hue="GenderId", palette="pastel")
    plt.title(f'Count Plot of {predictor} by Gender', fontsize=14)
    plt.xlabel(predictor, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.legend(title="Gender", loc='upper right')
    
```

# Insights 
- Females have marginally 2 credit cards 
- More female Europeans have taken credit cards.
- Female have owned more properties than male.

# Analysing with respect to Nationality 

```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality',
                                  'Occupation','Fee Structure','Loyalty Classification','Properties Owned',
                                  'Risk Weighting','Income Band']].columns):
    plt.figure(i, figsize=(10, 6))  # Set figure size
    sns.countplot(data=df, x=predictor, hue="Nationality", palette="pastel")
    plt.title(f'Count Plot of {predictor} by Nationality', fontsize=14)
    plt.xlabel(predictor, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.legend(title="Gender", loc='upper right')
```

```python
# HistPlot of value counts for different Ocuupation
sns.set_style("whitegrid")  # Clean background
sns.set_palette("Set2")     # Softer colors

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=False, bins=30, edgecolor='black')  # add bin edges
    plt.title(f'Histogram of {col}', fontsize=14, fontweight='bold')
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

# Numerical Analysis 

```python
numerical_cols= ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate analysis
plt.figure(figsize=(10,18))
for i, col in enumerate(numerical_cols):
    plt.subplot(9,1,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

# Insights
- All of them are right skewed

# Heatmaps
- To show correlation between all the attributes 

```python
numerical_cols= ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix= df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot= True, cmap='crest', fmt='.2f')
plt.title("Correlation Matrix")
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```

# Insights and Observation: 
- A high correlation between bank deposits and checking accounts and saving accounts is seen.
- Customers who have checking accounts are more likely to have savings accounts and vice versa.
- We can use this insights to segment customer based on their account types.
